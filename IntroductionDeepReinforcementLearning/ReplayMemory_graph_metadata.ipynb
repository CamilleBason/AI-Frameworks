# Lunar Launder

```python
fontsize=20
import pickle
mine = pickle.load(open("LunarLander/lunar_lander_metadata.pkl",'rb'))
n_mine = len(mine)
mine_rm = pickle.load(open("LunarLander/lunar_lander_RM_metadata.pkl",'rb'))
n_mine_rm = len(mine_rm)

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
sb.set_style("whitegrid")
fig = plt.figure(figsize=(25,12.5))
for i,(title,n,data) in enumerate([["WITHOUT",n_mine,mine], ["WITH",n_mine_rm,mine_rm]]):
    ax = fig.add_subplot(2,2,2*i+1)
    ax.plot(list(range(n)),[x[3] for x in data])
    ax.set_yticks(np.arange(-550,420,50))
    ax.set_xticks(np.arange(0,400,25))    
    ax.set_title("Score over Iteration %s Memory Replay" % title, fontsize=fontsize)
    plt.yticks(fontsize=12)
    plt.xticks(fontsize=12)
    ax.set_ylabel("Score", fontsize=16)
    if title=="WITH":
        ax.set_xlabel("Number of iteration", fontsize=14)
    
    ax = fig.add_subplot(2,2,2*i+2)
    ax.plot(list(range(n)),[x[2] for x in data])
    ax.set_yticks(np.arange(0,1050,50))
    ax.set_xticks(np.arange(0,420,25))
    ax.set_title(title)
    ax.set_title("Length of episode over Iteration %s Memory Replay" % title, fontsize=fontsize)
    plt.yticks(fontsize=12)
    plt.xticks(fontsize=12)
    ax.set_ylabel("Length of episode", fontsize=16)
    if title=="WITH":
        ax.set_xlabel("Number of iteration", fontsize=14)
plt.tight_layout()
```
